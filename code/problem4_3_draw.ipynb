```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('problem4_all.csv')

# 提取max_flow和delete_cnt列的数据
max_flow = data['max_flow']
delete_cnt = data['delete_cnt']

# 创建一个新的图表对象并设置标题
fig, ax = plt.subplots()
ax.set_title('Max Flow vs. Delete Count')

# 绘制散点图，以delete_cnt为横坐标，以max_flow为纵坐标
ax.plot(delete_cnt, max_flow)

# 添加坐标轴标签
ax.set_xlabel('Delete Count')
ax.set_ylabel('Max Flow')

# 显示图表
plt.show()
```
